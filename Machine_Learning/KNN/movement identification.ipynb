人类动作识别
步行，上楼，下楼，坐着，站立和躺着

数据采集每个人在腰部穿着智能手机，进行了六个活动（步行，上楼，下楼，坐着，站立和躺着）。采用嵌入式加速度计和陀螺仪，以50Hz的恒定速度捕获3轴线性加速度和3轴角速度，来获取数据

```python
import numpy as np

from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train = np.load('./man/x_train.npy')
y_train = np.load('./man/y_train.npy')

X_test = np.load('./man/x_test.npy')
y_test = np.load('./man/y_test.npy')
```

```python
X_train.shape

# 7352个样本，561个属性，人体动作
```

```python
X_test.shape

# 2947个测试数据，561个属性，人体动作
```

```python
%%time
knn = KNeighborsClassifier(n_neighbors=11)

knn.fit(X_train, y_train)

y_ = knn.predict(X_test)

print((y_ == y_test).mean())
```

```python
%%time
knn = KNeighborsClassifier()

knn.fit(X_train, y_train)

y_ = knn.predict(X_test)

print((y_ == y_test).mean())
```

```python
np.unique(y_test)
# 6种活动状态
```

```python
# 状态名称
labels = np.array(['walking','walk upstare','walk downstare','sitting','standing','lying'])
```

```python
# walking
np.argwhere(y_test == 1)

# walk upstare
np.argwhere(y_test == 2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12, 6))

for i in range(1, 7):
    
    index = np.argwhere(y_test == i)
    
    n = index[0, 0]
    
    # 两行三列,第i个图
    axes = plt.subplot(2, 3, i)
    axes.plot(X_test[n])
    
    axes.set_title(labels[i-1])
    
```
