```python
import numpy as np

import pandas as pd

from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split
```

```python
cancer = pd.read_csv('./data/cancer.csv', sep='\t')
cancer
```

```python
X = cancer.iloc[:, 2: ]
y = cancer['Diagnosis']
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
```

```python
# X_train中对应目标值y_train(M, B)
y_train.value_counts()
```

```python
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

y_ = knn.predict(X_test)

(y_ == y_test).mean()
```

```python
# 样本均衡化B(287) M(186) -------> 287
from imblearn.over_sampling import SMOTE
```

```python
# 该方法将数量比较少的一方,使用KNN近邻算法复制多份, 根据距离生成
smote = SMOTE()

smote.fit(X_train, y_train)
```

```python
# 生成均衡后的数据
X_train2, y_train2 = smote.fit_sample(X_train, y_train)
```

```python
X_train2.shape
```

```python
(y_train2 == 'B').sum()
```

```python
(y_train2 == 'M').sum()
```

```python
y_train.value_counts()
```

```python
knn = KNeighborsClassifier()

knn.fit(X_train2, y_train2)

# 测试的数据样本不均衡, 没有影响
y_ = knn.predict(X_test)

(y_ == y_test).mean()
```

```python
# 先进性归一化，train_test_split

v_min = X.min()
v_max = X.max()

X2 = (X - v_min)/(v_max - v_min)
```

```python
# 原数据
score = []
# 归一化之后数据准确率
score2 = []
for i in range(100):
    
#     前两个没有归一化，中间两个进行了归一化
    X_train,X_test,X_train2,X_test2,y_train,y_test = train_test_split(X,X2,y,test_size = 0.15)
    
    knn = KNeighborsClassifier()
    
    
    knn.fit(X_train,y_train)
    
    score.append(knn.score(X_test,y_test))
    
    knn.fit(X_train2,y_train)
    
    score2.append(knn.score(X_test2,y_test))
    
    
print('原数据平均准确率：%0.4f'%(np.mean(score)))

print('归一化平均准确率：%0.4f'%(np.mean(score2)))
```

```python
X2.std()
```

```python
# 另一种归一化Z-score归一化
v_mean = X.mean()

# 标准差
v_std = X.std()

# （V- min）/(max - min)
X3 = (X - v_mean)/v_std

# 特点X3，每个属性标准差都是1
X3.std()
```

```python
X3
```

```python
# 原数据
score = []
# 归一化之后数据准确率
score2 = []
for i in range(100):
    
#     前两个没有归一化，中间两个进行了归一化
    X_train,X_test,X_train3,X_test3,y_train,y_test = train_test_split(X,X3,y,test_size = 0.15)
    
    knn = KNeighborsClassifier()
    
    
    knn.fit(X_train,y_train)
    
    score.append(knn.score(X_test,y_test))
    
    knn.fit(X_train3,y_train)
    
    score2.append(knn.score(X_test3,y_test))
    
    
print('原数据平均准确率：%0.4f'%(np.mean(score)))

print('归一化平均准确率：%0.4f'%(np.mean(score2)))
```
