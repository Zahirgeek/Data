```python
import numpy as np

import pandas as pd

from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split
```

```python
cancer = pd.read_csv('./data/cancer.csv', sep='\t')
cancer
```

```python
X = cancer.iloc[:, 2: ]
y = cancer['Diagnosis']
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
```

```python
# X_trainä¸­å¯¹åº”ç›®æ ‡å€¼y_train(M, B)
y_train.value_counts()
```

```python
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

y_ = knn.predict(X_test)

(y_ == y_test).mean()
```

```python
# æ ·æœ¬å‡è¡¡åŒ–B(287) M(186) -------> 287
from imblearn.over_sampling import SMOTE
```

```python
# è¯¥æ–¹æ³•å°†æ•°é‡æ¯”è¾ƒå°‘çš„ä¸€æ–¹,ä½¿ç”¨KNNè¿‘é‚»ç®—æ³•å¤åˆ¶å¤šä»½, æ ¹æ®è·ç¦»ç”Ÿæˆ
smote = SMOTE()

smote.fit(X_train, y_train)
```

```python
# å…ˆè¿›æ€§å½’ä¸€åŒ–ï¼Œtrain_test_split

v_min = X.min()
v_max = X.max()

X2 = (X - v_min)/(v_max - v_min)
```

```python
# åŽŸæ•°æ®
score = []
# å½’ä¸€åŒ–ä¹‹åŽæ•°æ®å‡†ç¡®çŽ‡
score2 = []
for i in range(100):
    
#     å‰ä¸¤ä¸ªæ²¡æœ‰å½’ä¸€åŒ–ï¼Œä¸­é—´ä¸¤ä¸ªè¿›è¡Œäº†å½’ä¸€åŒ–
    X_train,X_test,X_train2,X_test2,y_train,y_test = train_test_split(X,X2,y,test_size = 0.15)
    
    knn = KNeighborsClassifier()
    
    
    knn.fit(X_train,y_train)
    
    score.append(knn.score(X_test,y_test))
    
    knn.fit(X_train2,y_train)
    
    score2.append(knn.score(X_test2,y_test))
    
    
print('åŽŸæ•°æ®å¹³å‡å‡†ç¡®çŽ‡ï¼š%0.4f'%(np.mean(score)))

print('å½’ä¸€åŒ–å¹³å‡å‡†ç¡®çŽ‡ï¼š%0.4f'%(np.mean(score2)))
```

```python
X2.std()
```

```python
# å¦ä¸€ç§å½’ä¸€åŒ–Z-scoreå½’ä¸€åŒ–
v_mean = X.mean()

# æ ‡å‡†å·®
v_std = X.std()

# ï¼ˆV- minï¼‰/(max - min)
X3 = (X - v_mean)/v_std

# ç‰¹ç‚¹X3ï¼Œæ¯ä¸ªå±žæ€§æ ‡å‡†å·®éƒ½æ˜¯1
X3.std()
```

```python
# åŽŸæ•°æ®
score = []
# å½’ä¸€åŒ–ä¹‹åŽæ•°æ®å‡†ç¡®çŽ‡
score2 = []
for i in range(100):
    
#     å‰ä¸¤ä¸ªæ²¡æœ‰å½’ä¸€åŒ–ï¼Œä¸­é—´ä¸¤ä¸ªè¿›è¡Œäº†å½’ä¸€åŒ–
    X_train,X_test,X_train3,X_test3,y_train,y_test = train_test_split(X,X3,y,test_size = 0.15)
    
    knn = KNeighborsClassifier()
    
    
    knn.fit(X_train,y_train)
    
    score.append(knn.score(X_test,y_test))
    
    knn.fit(X_train3,y_train)
    
    score2.append(knn.score(X_test3,y_test))
    
    
print('åŽŸæ•°æ®å¹³å‡å‡†ç¡®çŽ‡ï¼š%0.4f'%(np.mean(score)))

print('å½’ä¸€åŒ–å¹³å‡å‡†ç¡®çŽ‡ï¼š%0.4f'%(np.mean(score2)))
```
