```python
import numpy as np

from sklearn.neighbors import KNeighborsRegressor

import matplotlib.pylab as plt
%matplotlib inline
```

```python
# 80个点
# 预测正弦波
# 0~2pi 类比0~5000海里(每隔100公里获取风力数据, 50个)
# 问题是: 预测更具体的数据0, 1, 2, 3, 4, 5, 6... 现在需要分成5000份

# 第一步算法学习这50份数据规律

# 第二步算法, 预测5000份数据的值, 预测

X = np.linspace(0, 2*np.pi, 80)

y = np.sin(X)
plt.figure(figsize=(12, 9))
plt.scatter(X, y)
```

```python
knn = KNeighborsRegressor(n_neighbors=5, weights='distance')

# 需要将训练数据变成二维
knn.fit(X.reshape(-1, 1), y)

# 预测5000个数据
X_test = np.linspace(0, 2*np.pi, 5000).reshape(-1, 1)

# 一一对应
y_ = knn.predict(X_test)

y_.shape
```

```python
plt.figure(figsize=(12, 9))

plt.scatter(X_test, y_)
```
