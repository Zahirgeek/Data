```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = np.linspace(-5, 10, 100)
f = lambda x: (x - 2.5)**2 - 10
w = f(s)
# 导数
t = lambda x: 2*(x - 2.5)
plt.plot(s, w)
```

```python
# 求解函数最小值, 使用梯度下降, 一步一步找

# 随机给定结果
# size:输出元素的数量大小，可以是整数也可以是元组
result = np.random.randint(-5, 10, size=1)[0]
print(result)

step = 0.1
last = -6

results = []
# 精确度
p = 0.001

while(True):
    results.append(result)
    if np.abs(result - last) < p:
        break
    
    last = result
    result -= step*t(result)
    print(result)
    
plt.figure(figsize=(12, 9))
plt.plot(s, w)
ret = np.array(results)
# 绘制梯度下降经过的点
plt.scatter(ret, f(ret), marker='x', s=15, color='red')
```
