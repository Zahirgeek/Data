```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

from sklearn.cluster import KMeans

```

```python
bird = plt.imread('./data/bird_small.png')
bird.shape
```

```python
# 研究图片中的像素
# 宽高变成一维
X = bird.reshape(-1,3)
```

```python
# 图片中16384个像素
# 像素必然有相近
X.shape
```

```python
%%time
kmeans = KMeans(64)

kmeans.fit(X)

y_ = kmeans.predict(X)
```

```python
# 分成了64类，64个聚类中心------------>像素
# 每一个聚类中心表示颜色，代表这一类颜色
colors = kmeans.cluster_centers_
colors
```

```python
# 16384像素分类情况
y_
```
