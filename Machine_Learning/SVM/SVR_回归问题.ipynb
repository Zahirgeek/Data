```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 依然多种内核
from sklearn.svm import SVR
```

```python
X = np.linspace(0, 2*np.pi, 40).reshape(-1, 1)
y = np.sin(X)

plt.scatter(X, y)
```

```python
svr_linear = SVR(kernel='linear')

svr_rbf = SVR()

svr_poly = SVR(kernel='poly')

# 训练数据
svr_linear.fit(X, y)
svr_rbf.fit(X, y)
svr_poly.fit(X, y)
```

```python
# 测试数据
X_test = np.linspace(0, 2*np.pi, 2000).reshape(-1, 1)
```

```python
y_linear_ = svr_linear.predict(X_test)
y_rbf_ = svr_rbf.predict(X_test)
y_poly_ = svr_poly.predict(X_test)
```

```python
plt.figure(figsize=(9, 6))

plt.scatter(X, y)
plt.plot(X_test, y_linear_, label='linear')
plt.plot(X_test, y_rbf_, label='rbf')
plt.plot(X_test, y_poly_, label='poly')

plt.legend()
```

```python
svr_sigmoid = SVR(kernel='sigmoid')

svr_sigmoid.fit(X, y)

y_sigmoid_ = svr_sigmoid.predict(X_test)

plt.scatter(X, y)

plt.plot(X_test, y_sigmoid_)
```

```python
svr_sigmoid = SVR(kernel = 'sigmoid')

svr_sigmoid.fit(X,y)

y_sigmoid_ = svr_sigmoid.predict(X_test)

plt.scatter(X,y)

plt.plot(X_test,y_sigmoid_)
```
