```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.svm import SVC

import sklearn.datasets as datasets
```

```python
# 分为两类
X, y = datasets.make_blobs(centers=2)

plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# kernel内核:linear线性, rbf(基于半径), poly(多项式)
svc = SVC(kernel='linear')

svc.fit(X, y)
```

```python
# 100个样本, 每个样本2个属性
X.shape
```

```python
# 系数
w_ = svc.coef_
w_
```

```python
# 截距
b_ = svc.intercept_
b_
```

```python
# 方程对应三维空间中一条直线
z = f(x, y) = w_[0, 0]*x + w_[0, 1]*y + b_[0]
```

```python
plt.scatter(X[:, 0], X[:, 1],c=y)
```

```python
fig = plt.figure(figsize=(9, 6))

axes3D = Axes3D(fig)

axes3D.scatter(X[:, 0], X[:, 1], 0, c=y, cmap=plt.cm.PuOr)

x1 = np.linspace(-12, 1, 100)

y1 = np.linspace(-8, 6, 100)

z1 = w_[0, 0]*x1 + w_[0, 1]*y1 + b_[0]

axes3D.plot(x1, y1, z1, color='green')
```

```python
# 方程对应三维空间中一条直线
z = f(x, y) = w_[0, 0]*x + w_[0, 1]*y + b_[0]
# 设 z = 0, 三维空间变成二维空间

0 = w_[0, 0]*x + w_[0, 1]*y + b_[0]
w_[0, 1]*y = -w_[0, 0]*x - b_[0]
y = -w_[0, 0]/w_[0, 1]*x - 1/w_[0, 1]*b_[0]
```

```python
w1 = -w_[0, 0]/w_[0, 1]
w1
```

```python
b1 = - 1/w_[0, 1]*b_[0]
b1
```
