```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.svm import SVC
```

```python
# 对比三种内核 kernel
# 数据使用鸢尾花
import sklearn.datasets as datasets
```

```python
iris = datasets.load_iris()

# 二维, 第二维两个属性,画图显示
X = iris['data'][:, : 2]

y = iris['target']

plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
svc_linear = SVC(kernel='linear')

svc_rbf = SVC(kernel='rbf')

svc_poly = SVC(kernel='poly')

svc_linear.fit(X, y)

svc_rbf.fit(X, y)

svc_poly.fit(X, y)
```

```python
x1 = np.linspace(4, 8, 120)

y1 = np.linspace(2, 4.5, 80)

# 生成网格点坐标矩阵
X1, Y1 = np.meshgrid(x1, y1)

X_test = np.concatenate([X1.reshape(-1, 1), Y1.reshape(-1, 1)], axis=1)
X_test.shape
```

```python
y_linear_ = svc_linear.predict(X_test)

y_rbf_ = svc_rbf.predict(X_test)

y_poly_ = svc_poly.predict(X_test)

# 绘制图像

plt.figure(figsize=(16, 10))

axes = plt.subplot(2, 2, 1)
axes.scatter(X[:, 0], X[:, 1], c=y)

axes = plt.subplot(2, 2, 2)
# axes.scatter(X_test[:, 0], X_test[:, 1], c=y_linear_)
# 轮廓面, 图像更加平滑
axes.contourf(X1, Y1, y_linear_.reshape(80, 120))
axes.set_title('Linear')

axes = plt.subplot(2, 2, 3)
axes.scatter(X_test[:, 0], X_test[:, 1], c=y_rbf_)
axes.set_title('rbf')

axes = plt.subplot(2, 2, 4)
axes.scatter(X_test[:, 0], X_test[:, 1], c=y_poly_)
axes.set_title('poly')
```
